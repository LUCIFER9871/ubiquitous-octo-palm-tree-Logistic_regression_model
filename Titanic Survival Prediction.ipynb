## Titanic Survival Prediction

## Import Library

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Import Dataset and preprocessing

```python
df = pd.read_csv('Titanic-Dataset.csv')
df=pd.DataFrame(df)
```

```python
df.head()
```

```python
df['Parch'].nunique()
```

```python
df['Fare'].nunique()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])
```

```python
df.info()
```

```python
df['Fare']
```

```python
df = df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis = 1)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Age'].isna().sum()
```

```python
df= pd.get_dummies(df, columns=['Sex', 'Embarked', 'Pclass', 'SibSp', 'Parch'])
```

```python
df
```

```python
df.nunique()
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
df= df.drop(['Embarked_C','Pclass_1' ,'SibSp_0' , 'Parch_0'] , axis =1)
```

```python
df.head()
```

```python
df=df.replace(to_replace=True, value=1)
```

```python
df
```

```python
df=df.replace(to_replace=False, value=0)
```

```python
df 
```

## Train Test Split

```python
y= df['Survived']
```

```python
y.head()
```

```python
X = df.iloc[:, 1:]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test , y_train, y_test = train_test_split(X , y, test_size = 0.20 , random_state = 42)
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
y_train.shape
```

```python
X_train.shape
```

## Model Training 

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=5000)
```

```python
model_1=model.fit(X_train, y_train)
```

```python
model_1
```

```python
y_pred = model_1.predict(X_test)
```

```python
y_pred
```

## Model Evaluation and Accuracy

```python
from sklearn.metrics import confusion_matrix , precision_score, recall_score , f1_score , classification_report
```

```python
cm = confusion_matrix(y_test , y_pred)
```

```python
print("Confusion Matrix :\n ",cm)
```

```python
print('Recall:', recall_score(y_test, y_pred))
```

```python
print('Precision:' , precision_score(y_test , y_pred))
```

```python
print('F1 Score: ' , f1_score(y_test , y_pred))
```

```python
print('\n Classification Report :\n', classification_report(y_test ,y_pred))
```
